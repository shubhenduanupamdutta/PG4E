## Example Codes for Using JSON serialization and deserialization in Python

### Using `json.dumps` to serialize a Python object to a JSON string


```python
import json

data = {}
data["name"] = "Chuck"
data["phone"] = {}
data["phone"]["type"] = "intl"
data["phone"]["number"] = "+1 734 303 4456"
data["email"] = {}
data["email"]["hide"] = "yes"

## Serialize the data to a JSON string
print(json.dumps(data, indent=4))
```

### Using `json.loads` to deserialize a JSON string back to a Python object


```python
data = """
{
    "name": "Chuck",
    "phone": {
        "type": "intl",
        "number": "+1 734 303 4456"
    },
    "email": {
        "hide": "yes"
    }
}
"""
print("Type of data:", type(data))
## Deserialize the JSON string back to a Python object
info = json.loads(data)
print("Type of info:", type(info))
print(f"data as python object:\n{info = }")
```
